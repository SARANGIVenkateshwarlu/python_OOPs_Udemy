ðŸ”¹ 2. Static Method â€” Definition

```python
# A static method is a method that:
# - Does NOT use the 'self' parameter
# - Does NOT access instance attributes
# - Belongs to the class logically, not the object
#   Defined using the @staticmethod decorator
#   Used for utility / helper functionality
```

ðŸ”¹ 3. Decorator â€” Definition


```python
# A decorator is a special function that:
# - Modifies or extends the behavior of another function
# - Is written using the '@' symbol

# @staticmethod  converts a normal method into a static method
```

ðŸ”¹ 4. Difference Between Instance Method & Static Method

```python
# Instance Method:
# - Uses self
# - Accesses object data
# - Needs an object to be called

# Static Method:
# - No self
# - No object data
# - Can be called using class or object
```

âœ… EXAMPLE 1: Instance Method (Used Inside a Loop)

```python
# --------------------------------------------------
# EXAMPLE 1: INSTANCE METHOD
# --------------------------------------------------

class Employee:
    # Instance method (uses self)
    def employeeDetails(self):
        # Instance attribute
        self.name = "Ben"

        print("Employee Name:", self.name)


# Creating object and invoking emply details
employee = Employee()

# Each call uses the object and self
employee.employeeDetails()
```

âœ… Explanation python




```python
# employeeDetails() is an instance method
# It uses self to store and access data
# The method is invoked multiple times using a loop
```

âœ… EXAMPLE 2: Static Method (Used Inside a Loop)

```python
# --------------------------------------------------
# EXAMPLE 2: STATIC METHOD
# --------------------------------------------------

class Company:

    @staticmethod
    def welcomeMessage():
        # Static method does NOT use self
        print("Welcome to our organization!")


# Calling static method and Can be called using class name
Company.welcomeMessage()
```

```python
# welcomeMessage() is a static method
# It does not depend on object data, binding has been ingnore
# No self parameter is required
# Used for common messages or utilities
```

âœ… When to Use What?

```python
# Use Instance Method:
# - When method needs object data
# - When attributes are involved
# Use Static Method:
# - When method does not need object data
# - For helper, utility, or display logic
```
